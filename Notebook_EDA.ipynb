# creation de notre baseline model

```python

immo_df_snan = immo_df_snan.rename(columns={'housing_median_age':'Age_median_quartier','households':'nb_menages'})
```

```python
immo_df_snan.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
X=pd.DataFrame(immo_df_snan[['Age_median_quartier', 'total_rooms', 'total_bedrooms', 'population', 'nb_menages', 'median_income','ocean_proximity']], 
               columns = ['Age_median_quartier', 'total_rooms', 'total_bedrooms', 'population', 'nb_menages', 'median_income'])
Y = immo_df_snan['median_house_value']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.33, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
lmodellineaire = LinearRegression()
lmodellineaire.fit(X_train, Y_train)
```

```python
y_train_predict = lmodellineaire.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)

print('La performance du modle sur la base dapprentissage')
print('--------------------------------------')
print('Lerreur quadratique moyenne est {}'.format(rmse))
print('le score R2 est {}'.format(r2))
print('\n')
```

```python
y_test_predict = lmodellineaire.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)
 
print('La performance du modle sur la base de test')
print('--------------------------------------')
print('Lerreur quadratique moyenne est {}'.format(rmse))
print('le score R2 est {}'.format(r2))
```

## par rapport au median income et a la ocean_proximity

```python
XX=pd.DataFrame(immo_df_snan[['median_income','ocean_proximity']], 
               columns = ['median_income'])
YY = immo_df_snan['median_house_value']

XX_train, XX_test, YY_train, YY_test = train_test_split(XX, YY, test_size = 0.33, random_state=5)
print(XX_train.shape)
print(XX_test.shape)
print(YY_train.shape)
print(YY_test.shape)
```

```python
lmodellineaire = LinearRegression()
lmodellineaire.fit(XX_train, YY_train)
```

```python
yy_train_predict = lmodellineaire.predict(XX_train)
rmse = (np.sqrt(mean_squared_error(YY_train, yy_train_predict)))
r2 = r2_score(YY_train, yy_train_predict)

print('La performance du modle sur la base dapprentissage')
print('--------------------------------------')
print('Lerreur quadratique moyenne est {}'.format(rmse))
print('le score R2 est {}'.format(r2))
print('\n')
```

```python
yy_test_predict = lmodellineaire.predict(XX_test)
rmse = (np.sqrt(mean_squared_error(YY_test, yy_test_predict)))
r2 = r2_score(YY_test, yy_test_predict)
 
print('La performance du modle sur la base de test')
print('--------------------------------------')
print('Lerreur quadratique moyenne est {}'.format(rmse))
print('le score R2 est {}'.format(r2))
```
